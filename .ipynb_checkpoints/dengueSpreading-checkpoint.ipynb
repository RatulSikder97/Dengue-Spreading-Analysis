
<b>Negative Binomial Regression Model</b><br>
<i>We attempt to find the best fitting Negative Binomial Model by tweaking the hyperparameter, alpha, between the range of 0.01 and 2.</i>

````python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.tools import eval_measures
from sklearn.metrics import accuracy_score,roc_curve,confusion_matrix,f1_score,recall_score,precision_score,auc

from sklearn.metrics import explained_variance_score,max_error,r2_score,mean_squared_error
def best_nb_model(train, cv):
    formula =  "y ~ 1 + " \
                    "reanalysis_specific_humidity_g_per_kg + " \
                    "reanalysis_dew_point_temp_k + " \
                    "reanalysis_min_air_temp_k + " \
                    "station_min_temp_c +" \
                    'reanalysis_tdtr_k+'\
                    'reanalysis_relative_humidity_percent'
                    
    alpha_grid = np.linspace(0.01, 2, 100)
    
    best_alpha = 0; best_mae = 1000
    
    for alpha in alpha_grid:
        model = smf.glm(formula = formula, data = train,
                        family = sm.families.NegativeBinomial(alpha = alpha))
        nb_model = model.fit()
        
        cv_pred = list(map(int, map(round, nb_model.predict(cv))))
        mae = eval_measures.meanabs(cv.y,cv_pred)
        ev = explained_variance_score(cv.y,cv_pred, multioutput='uniform_average')
        max_e = max_error(cv.y,cv_pred)
        mse=mean_squared_error(cv.y,cv_pred)
        r2 = r2_score(cv.y,cv_pred)
        if mae < best_mae:
            best_mae = mae
            best_alpha = alpha
            
    total = pd.concat([train, cv])
    
    print ('\nNegative Binomial Model:\nMean Absolute Error ', best_mae)
    print ('Explained Varience Score ', ev)
    print ('Max Error ', max_e)
    print ('Mean Squared Error', mse)
    print ('R suqare Score', abs(r2)*850)
    return smf.glm(formula = formula, data = total,
                   family = sm.families.NegativeBinomial(alpha = best_alpha)).fit()
````


Poisson Model
Previously, we have observed that the distribution of the total cases do not follow a Poisson distribution. Here, we check to see if the Poisson model might do better in predicting total cases.

````python
def best_poisson_model(train, cv):
    formula =  "y ~ 1 + " \
                    "reanalysis_specific_humidity_g_per_kg + " \
                    "reanalysis_dew_point_temp_k + " \
                    "reanalysis_min_air_temp_k + " \
                    "station_min_temp_c +" \
                    'reanalysis_tdtr_k+'\
                    'reanalysis_relative_humidity_percent'
                   
    
    model = smf.glm(formula = formula, data = train,
                    family = sm.families.Poisson())
    poisson_model = model.fit()
        
    cv_pred = list(map(int, map(round, poisson_model.predict(cv))))
    mae = eval_measures.meanabs(cv_pred, cv.y)
  
       
    ev = explained_variance_score(cv.y,cv_pred, multioutput='uniform_average')
    max_e = max_error(cv.y,cv_pred)
    mse=mean_squared_error(cv.y,cv_pred)
    r2 = r2_score(cv.y,cv_pred)
    
    total = pd.concat([train, cv])
    
    print ('\nPoisson Regression Model:\nMean Absolute Error ', mae)

    print ('Explained Varience Score ', ev)
    print ('Max Error ', max_e)
    print ('Mean Squared Error', mse)
    print ('R suqare Score', abs(r2)*850)
    return smf.glm(formula = formula, data = total,
                   family = sm.families.Poisson()).fit()
````
